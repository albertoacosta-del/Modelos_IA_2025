```python
# üì¶ Instalaci√≥n de librer√≠as necesarias
!pip install -q pandas seaborn matplotlib

# üîß Importaci√≥n de m√≥dulos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# üéØ Configuraci√≥n de estilo para gr√°ficos
sns.set(style="whitegrid")
```

```python
# üìÅ Carga manual del archivo CSV
from google.colab import files

print("üîº Por favor, sube tu archivo CSV (por ejemplo: train.csv)")
uploaded = files.upload()

# üóÇÔ∏è Lectura del archivo
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# üìä Informaci√≥n b√°sica del dataset
print(f"‚úÖ Dimensiones del archivo: {df.shape}")
display(df.head())
```

```python
# üßæ Detalles de columnas y tipos de datos
df.info()
```

```python
# üìà Distribuci√≥n de la variable objetivo (si existe)
target_col = "RENDIMIENTO_GLOBAL"
if target_col in df.columns:
    print(f"\nüìä Distribuci√≥n de clases en '{target_col}':")
    print(df[target_col].value_counts())

    # üìâ Visualizaci√≥n de la distribuci√≥n
    plt.figure(figsize=(6,4))
    sns.countplot(y=target_col, data=df, order=df[target_col].value_counts().index, palette="coolwarm")
    plt.title("Distribuci√≥n de niveles de rendimiento")
    plt.xlabel("Cantidad de estudiantes")
    plt.ylabel("Nivel")
    plt.tight_layout()
    plt.show()
else:
    print(f"‚ö†Ô∏è La columna '{target_col}' no se encuentra en el archivo.")
```

```python
# üó∫Ô∏è An√°lisis por departamento (si existe)
localidad_col = "E_PRGM_DEPARTAMENTO"
if localidad_col in df.columns:
    plt.figure(figsize=(8,5))
    top_deptos = df[localidad_col].value_counts().nlargest(10)
    sns.barplot(x=top_deptos.values, y=top_deptos.index, palette="viridis")
    plt.title("Top 10 departamentos por n√∫mero de estudiantes")
    plt.xlabel("Estudiantes")
    plt.ylabel("Departamento")
    plt.tight_layout()
    plt.show()
else:
    print(f"‚ö†Ô∏è La columna '{localidad_col}' no est√° disponible.")
```
